```python
import matplotlib.pyplot as plt 
from env import simulator
```

```python
N = 10000
T = 100
match_prob = [[0.8, 0.1], [0.8, 0.1]]
ratio1 = 0.5
ratio2 = 0.5
initial_guess = [0.5, 0.5]
r_s, r_o = simulator(N, T, match_prob, ratio1, ratio2, initial_guess)
```

```python
r_s_ = [i/N for i in r_s]
r_o_ = [i/N for i in r_o]
plt.plot([i for i in range(T)], r_s_)
plt.plot([i for i in range(T)], r_o_)
```
