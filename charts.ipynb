```python
import matplotlib.pyplot as plt 
from simplified import *
```

```python
N = 10000
T = 200
match_prob = [[0.8, 0.1], [0.8, 0.1]]
ratio1 = 0.8
ratio2 = 0.2
initial_guess = [ratio1, ratio2]
r_s, r_o, r_r, _= simulation(N, T, match_prob, ratio1, ratio2, initial_guess)

partial_s, partial_o, partial_r, partial_e = simulation(N, T, match_prob, ratio1, ratio2, initial_guess, full_info=False)
```

```python
plt.figure(figsize=(10, 8))
r_s_ = [i/N for i in r_s]
r_o_ = [i/N for i in r_o]
r_r_ = [i/N for i in r_r]
# r_e_ = [i/N for i in r_e]
p_s_ = [i/N for i in partial_s]
p_e_ = [i/N for i in partial_e]
plt.plot([i for i in range(T)], r_s_, label='greedy')
plt.plot([i for i in range(T)], r_o_, label='offline')
plt.plot([i for i in range(T)], r_r_, label='random')
# plt.plot([i for i in range(T)], r_e_, label='epsilon-greedy')
plt.plot([i for i in range(T)], p_s_, label='partial_info_greedy')
plt.plot([i for i in range(T)], p_e_, label='partial_info_e_greedy')

plt.xlabel('Iteration')
plt.ylabel('match rate')
plt.title(f'{match_prob}')
plt.legend()
plt.show()
```
